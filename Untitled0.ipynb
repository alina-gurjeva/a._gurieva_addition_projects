# Условие задачи

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Вам предстоит построить модель, которая предскажет температуру стали.

### Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `data_arc.csv` — данные об электродах;
- `data_bulk.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time.csv` *—* данные о подаче сыпучих материалов (время);
- `data_gas.csv` — данные о продувке сплава газом;
- `data_temp.csv` — результаты измерения температуры;
- `data_wire.csv` — данные о проволочных материалах (объём);
- `data_wire_time.csv` — данные о проволочных материалах (время).

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

3. Выберем baseline модель. Например, линейную регрессию, поскольку она не требует многих настроек гиперпараметров. Выясним опытным путем, на каких сочетаниях признаков модель даст наилучший результат.

```
def splitdata(data, target='last_temp'):

    '''Функция выделит тестовый столбец из данных, а также разобьет все данные
    на валидационную, тестовую и обучающую выборки. Вернет 8 датасетов 
    (включая более полные данные для обучения модели перед финальным тестирование) '''
    
    from sklearn.model_selection import train_test_split
    train_valid, test = train_test_split(data, test_size=0.2, random_state = 88184)
    train, valid = train_test_split(train_valid, test_size=0.2, random_state = 88184)

    def targetit(data):
        x, y = data.drop(target, axis=1), data[target]
        return x, y

    train_validX, train_validY = targetit(train_valid)
    trainX, trainY = targetit(train)
    testX, testY = targetit(test)
    validX, validY = targetit(valid)

    return train_validX, train_validY, trainX, trainY, validX, validY, testX, testY
```

```
def testit(trainX, trainY, validX, validY):
    '''Функция возьмет обучающую и валидационную выборку и определит МАЕ линейной 
    регрессии. Поможет опредеделить, какие признаки необходимо оставить в данных.'''
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import mean_absolute_error
    model = LinearRegression()
    model.fit(trainX, trainY)
    prediction = model.predict(validX)
    scoremae = mean_absolute_error(prediction, validY)
    return scoremae
```

```
def make_PCAit(dfX,dfV,K):
    from sklearn.decomposition import PCA
    pca = PCA(n_components = K)
    f = pca.fit(dfX)
    fX = f.transform(dfX)
    fV = f.transform(dfV)
    dfXn = pd.DataFrame(fX).set_index(dfX.index)
    dfVn = pd.DataFrame(fV).set_index(dfV.index)
    return dfXn, dfVn

```

```
def tryK():
    Klist = []
    maelist = []
    for k in range(10, 30):
        dfX, dfV = make_PCAit(trainX, validX,k)
        Klist.append(k)
        sc = testit(dfX, trainY, dfV, validY)
        maelist.append(sc)
    result = pd.DataFrame()
    result['k'] = Klist
    result['mae'] = maelist
    return result 
```

```
def BestSkl(params):
    '''Функция примет на вход параметры для grid search случайного леса. 
    Вернет обученную модель, датасет с предсказаниями модели и наилучший достигнутый МАЕ'''
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.model_selection import GridSearchCV
    from sklearn.metrics import mean_absolute_error
    model = RandomForestRegressor(random_state = 6)
    grid = GridSearchCV(model, params, cv=5, scoring = 'neg_mean_absolute_error')
    grid.fit(trainX, trainY)
    model = grid.best_estimator_
    model.fit(trainX, trainY)
    prediction = model.predict(validX)
    scorem = mean_absolute_error(prediction, validY)
    return model, prediction, scorem

```

```
def BestLight(params):
    '''Функция примет на вход параметры для lightgbm.
    Вернет обученную модель, датасет с предсказаниями модели и наилучший достигнутый МАЕ'''
    import lightgbm as lgb
    from sklearn.model_selection import GridSearchCV
    from sklearn.metrics import mean_absolute_error
    model = lgb.LGBMRegressor(random_state = 6, iterations = 300, boosting_type = 'goss')
    grid = GridSearchCV(model, params, cv=5, scoring='neg_mean_absolute_error')
    grid.fit(trainX, trainY)
    model = grid.best_estimator_
    model.fit(trainX, trainY)
    pred = model.predict(validX)
    sc = mean_absolute_error(validY, pred)
    return model, pred, sc
```

```
def BestCat(params):
    '''Функция примет на вход параметры для lightgbm.
    Вернет обученную модель, датасет с предсказаниями модели и наилучший достигнутый МАЕ'''
    from catboost import CatBoostRegressor
    from sklearn.metrics import mean_absolute_error
    model = CatBoostRegressor(loss_function = "MAE", random_state=6, iterations=1000)
    model.grid_search(params, X = trainX, y = trainY, plot=True, 
                               search_by_train_test_split = True, cv=5, train_size=0.8, 
                               shuffle = True, verbose=True, refit=True)
    
    pred = model.predict(validX)
    sc = mean_absolute_error(validY, pred)
    return model, pred, sc
```

Возьмем получившиеся модели с наилучшими для них гиперпараметрами и напишем функцию, которая будет предсказывать результат, как средний по итогам голосования всех 5 моделей. 

```
def makepred(X):
    '''Функция забирает изначальные обучающие признаки и возвращает 
    датасет из предсказаний 4х моделей.'''

    lgbdt = pd.Series(modelgbdt.predict(X))
    gossm = pd.Series(model.predict(X))
    forest = pd.Series(forestmodel.predict(X))
    cat = pd.Series(catmodel.predict(X))
    df = pd.DataFrame()
 
    df['lgbdt'] = lgbdt
    df['gossm'] = gossm
    df['forest'] = forest
    df['cat'] = cat
    df = df.set_index(X.index)
    return df
```

Это дает основу для предположения, что усредненные предсказания моделей могут быть ближе к истине, чем предсказания отдельных моделей.


Напишем функцию, которая будет возвращать список с усредненными предсказаниями по моделям. 

```
def meanpred(X):
    df = makepred(X)
    df['pred'] = (df['lgbdt']+df['gossm']+df['forest']+df['cat'])/4
    return df['pred']
```

```
def meanpredfinal(X):
    '''Функция обучает модели на более полных данных,
    поскольку будет предсказывать значения тестовой выборки. 
    Возвращает среднее от предсказаний 4-х моделей. '''
    model.fit(train_validX, train_validY)
    catmodel.fit(train_validX, train_validY)
    forestmodel.fit(train_validX, train_validY)
    modelgbdt.fit(train_validX, train_validY)    
    df = makepred(X)
    df['pred'] = (df['lgbdt']+df['gossm']+df['forest']+df['cat'])/4
    return df['pred']
```

# Вывод по проекту

Нам были дана информация:

    data_arc.csv — данные об электродах;
    data_bulk.csv — данные о подаче сыпучих материалов (объём);

    data_gas.csv — данные о продувке сплава газом;
    data_temp.csv — результаты измерения температуры;
    data_wire.csv — данные о проволочных материалах (объём);
    data_wire_time.csv — данные о проволочных материалах (время).


Задача была - построить модель, которая предскажет температуру стали.

В первом разделе мы ознакомились с данными, предобработали их (очистили от выбросовых и нулевых значений), и проанализировали.

По итогам мы получили датасет alldata, где собрали все предположительно нужные для обучения признаки. 

Во втором разделе мы:

-    Выделили целевой признак и разделили выборку на 3 части: тестовую, валидационную и обучающую.
-    Протестировали различные подходы к отбору значимых признаков из полных данных. Выяснили, на каких сочетаниях признаков линейная модель дает наилучший результат.
-    Наилучшее сочетание признаков проверили на необходимость стандартизации/нормализации.
-    Сформировали итоговый датасет для дальнейшего обучения более сложных моделей.

В итоговом датасете оказались 5 признаков для обучения:

- 'first_temp' - первичная температура, 
- 'seconds' - время работы в секундах,
- 'fullPowerSum' - суммированная полная мощность, 
- 'AfullSum' - суммированная полная работа,
- 'gasMean' - усредненное по партии количество газа.
